### 특정 날짜 전체 기사

```python
url = "https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=020&date=20221206&page=100"
bs = bs_from_url(url)
last_page_num = int(bs.select('div.paging > strong')[0].text.strip())
last_page_num
```

```python
def bs_from_url(url) :
    respone = requests.get(url, 
        headers={
            'User-Agent': 'Mozilla 5.0'
        }
    )
    bs = BeautifulSoup(respone.text,"lxml")
    return bs

def whole_url_title(urls_and_titles):
    titles = []
    urls = []
    for url_and_title in urls_and_titles :
        if len(url_and_title) == 1:
            title = url_and_title.text.strip()
            url = url_and_title.get("href").strip()
            titles.append(title)
            urls.append(url)
        else :
            continue
    return titles,urls

def get_contents_from_url(url):
    bs = bs_from_url(url)

    if bs.select('#contents'):
        return bs_from_url(url).select('#contents')[0].text.strip()

    if bs.select('.content'):
        return bs_from_url(url).select('.content')[0].text.strip()
    
    
```

```python
donga = pd.DataFrame()

for date in tqdm(pd.date_range('20221205','20221206')):
    date = date.strftime("%Y%m%d")
    for i in range (1,last_page_num+1):
        url = f"https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=020&date={date}&page={i}"
        bs = bs_from_url(url)
        urls_and_titles = bs.select(".list_body.newsflash_body > ul dt > a")
        titles, urls = whole_url_title(urls_and_titles)
        contents = [get_contents_from_url(url) for url in urls]
        
        temp = pd.DataFrame({
            'date' : '20021206',
            'press_name' : '동아일보',
            'content' : contents,
            'title' : titles,
            'url' : urls 
        })
        donga = donga.append(temp)

donga
```
